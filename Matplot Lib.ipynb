```python
import numpy as np
# x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2
# plt.plot(x,y, color="black", marker="o")
plt.plot(x,y, color="black", linewidth=5, label="xˆ3")
plt.plot(x,y2,color="red", label="xˆ2")

plt.axis([0,10,0,200])
plt.ylabel("y")
plt.xlabel("x")
# plt.grid()

plt.text(2,80,"Text")
plt.legend()
plt.title("MatplotLib")
plt.show()
```

```python
# Bubble Chart 
import numpy as np
year = np.arange(2012,2018,1)
salary = np.array([12,13,14,17,19,20])
population = np.array([100,120,180,250,300,370])
print(year)
print(salary)
print(population)
plt.scatter(year,salary,s=population, alpha = 0.5)
plt.title("Bubble Chart")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,s=population,c = c, alpha = 0.5, edgecolor="b", marker='^')
plt.title("Bubble Chart")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c = c, alpha = 0.5, edgecolor="b", marker='^')
plt.title("Bubble Chart")
plt.xlabel("x")
plt.ylabel("y")
plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.show()
```

```python
# pie graphs
sizes = [3,4,6,2]
colors = ["blue","purple","red","pink"]
labels= ["A","B","C","D"]
explode = [0.1,0.2,0,0]
plt.pie(sizes,colors = colors,explode = explode,labels=labels, autopct="%.2f", counterclock=False, startangle=100)
plt.axis("equal")
plt.show()
```

```python
a = [1,2,3,4,1,3,4,2,1,1]
plt.hist(a)
plt.show()
```

```python
xt = np.arange(23)
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
plt.hist(b,bins=21, edgecolor="black") # [a,b] b not included but in last bin it is included.
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b,bins=[1,3,6,10,15,20])
plt.axis([0,25,0,10])
plt.show()
```

```python
# bar graphs
year = np.arange(2012,2018,1)
salary = np.array([12,13,14,17,19,20])
population = np.array([100,120,180,250,300,370])
plt.bar(year,population)
plt.show()
```

```python
plt.bar(year,population)
plt.xticks(rotation=40)
plt.show()
```
