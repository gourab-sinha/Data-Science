```python
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
i = 0
driver.back()
adj_btn = driver.find_element_by_link_text('Board of Accountancy')
adj_btn.click()
dataFrameList = []
while True:
    try:
        job_id = 'MainContent_uxLevel2_JobTitles_uxJobTitleBtn_'+str(i)
        job_btn = driver.find_element_by_id(job_id)
        job_btn.click()
        data = BeautifulSoup(driver.page_source,'lxml')
        emp_table = data.table
        df = pd.read_html(str(emp_table),header = 0)
        dataFrameList.append(df[0])
        driver.back()
        i+=1
    except NoSuchElementException:
        break
        
        
dataFrameList
```

```python
from selenium.common.exceptions import NoSuchElementException


a = driver.find_element_by_link_text('Board of Accountancy')
a.click()

dataFrameList = []

i = 0
while True : 
    try : 
        link = driver.find_element_by_id('MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
        link.click()

        data = BeautifulSoup(driver.page_source, 'lxml')
        emp_table = data.table
#         print(type(emp_table))
        df = pd.read_html(str(emp_table), header = 0)
        dataFrameList.append(df[0])
        driver.back()
        i += 1
    except NoSuchElementException :
        break
    
dataFrameList
```

```python
# Four agencies job details
list_job_board = ['Abstracters Board of Examiners', 'Adjutant General',
                  'Attorney General','Bank Commissioner']

dataFrameList = []
for board in list_job_board:
    board_btn = driver.find_element_by_link_text(board)
    board_btn.click()

    i = 0
    while True : 
        try : 
            link = driver.find_element_by_id('MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(i))
            link.click()

            data = BeautifulSoup(driver.page_source, 'lxml')
            emp_table = data.table
    #         print(type(emp_table))
            df = pd.read_html(str(emp_table), header = 0)
            dataFrameList.append(df[0])
            driver.back()
            i += 1
        except NoSuchElementException :
            break
    driver.back()

dataFrameList
    
```

```python
pd.concat(dataFrameList,ignore_index = True)
```
